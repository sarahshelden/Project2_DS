```python
# import dependencies
import pandas as pd
import numpy as np
import requests
import json
from pandas.io.json import json_normalize

from datetime import date

today = str(date.today())

listx = ["33.972386,-84.231986","47.6062,-122.3321", "42.3601,-71.0589","34.0522,-118.2437"]
# listname = ["Norcross", "Seattle", "Boston", "Los Angeles"]


dflist = pd.DataFrame()

for i in listx:
    url = f"https://api.darksky.net/forecast/fb0e09b8e9092cef1008bfe2eef00118/{i}?extend=hourly"
    #print(url)
    response = requests.get(url)
    data = response.json()
    citydf = json_normalize(data['hourly']['data'])
    df = citydf.loc[:,["time","temperature","pressure","humidity"]]
    df['clean time'] = pd.to_datetime(df['time'], unit='s')
    df = df[["clean time", "temperature", "pressure","humidity"]]
    df['city'] = i
    df['run_date'] = today
    dflist = dflist.append(df)

print(dflist)
```

```python

with open('sevendayforecast.csv', 'a') as f:
    dflist.to_csv(f, header=False)
```
