```python
# import dependencies
import pandas as pd
import numpy as np
import requests
import json
from pandas.io.json import json_normalize
```

```python
# Retrive Hourly Data
url = "https://api.darksky.net/forecast/fb0e09b8e9092cef1008bfe2eef00118/33.972386,-84.231986"
response = requests.get(url)
data = response.json()
complete = json_normalize(data['hourly']['data'])
```

```python
# Isolate data we need
df = complete.loc[:,["time","temperature","pressure","humidity"]]
df.head()
```

```python
df['clean time'] = pd.to_datetime(df['time'], unit='s')
df.head()
```

```python
df = df[["clean time", "temperature", "pressure","humidity"]]
df.head()
```

```python
# Save to Csv
df.to_csv("FORCASTING.csv", sep=',', encoding='utf-8')
```
